```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
df=data
```

```python
print("Welcome to Abba's renowned data analysis. Ensure your input file is readily available")
```

```python
df=df.drop_duplicates()
```

### We need to get rid of any duplicates or null values that may be present in the dataset. Null values can be handled in a variety of ways, including dropping the rows or columns containing null values or filling them with the mean or median of the corresponding column. The approach taken depends on the data and analysis being run. We can easily get rid of duplicates by using the drop duplicates() function.

```python
df=df.dropna()
```

```python
print("General statistics of your data in its entity:")
print(df.describe())
```

```python
print("Statistics with respect to columns:")
```

```python
print(df[['Alcohol','Volatile Acidity','Citric Acid','Residual Sugar','Chlorides',
         'Density','pH','Sulphates','Quality']].describe())
```

```python
print("\nHidden Inferences:")
print("The highest rated wine has an alcohol content of",df['Alcohol'].max())
print("The lowest rated wine has an alcohol content of",df['Alcohol'].min())
print("The average rating for all wine is", df['Quality'].mean())
print("The median rating for all wine is", df['Quality'].median())
```

```python
sns.countplot(x='Quality', data=df)
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
print("Relationships with regards to quality")
sns.boxplot(x='Quality',y='Alcohol',data=df)
plt.title('Relationship between Alcohol Content and Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
sns.boxplot(x='Quality', y='Residual Sugar', data=df)
plt.title('Relationship between Residual Sugar and Quality')
plt.xlabel('Quality')
plt.ylabel('Residual Sugar')
plt.show()
```

```python
sns.scatterplot(x='Fixed Acidity', y='pH', hue='Quality', data=df)
plt.title('Relationship between Acidity and Quality')
plt.xlabel('Fixed Acidity')
plt.ylabel('pH')
plt.show()
```

```python
sns.scatterplot(x='Sulphates', y='Alcohol', hue='Quality', data=df)
plt.title('Relationship between Sulphates and Quality')
plt.xlabel('Sulphates')
plt.ylabel('Alcohol Content')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation (Between Columns) Heatmap")
plt.show()
```

```python
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```
